```python
from abc import ABC, abstractmethod

# ----------------Abstraction-----------------
class Person(ABC):
  @abstractmethod
  def show_details(self):
    pass

# ---------------Inheritance and Encapsulation----------
class Student(Person):
  count = 0
  def __init__(self,roll,name,marks):
    self.__roll = roll
    self.__name = name
    self.__marks = marks
    Student.count +=1

  # getter
  @property
  def marks(self):
    return self.__marks

  # setter
  @marks.setter
  def marks(self,value):
      if value >= 0 and value <=100:
         self.__marks= value
      else:
          raise ValueError("Marks must be positive")

  # method overriding
  def show_details(self):
    return f"Roll: {self.__roll}, Name: {self.__name}, Marks: {self.__marks}"

  def __add__(self,other):
    return self.__marks + other.__marks

  # static method
  @staticmethod
  def greet():
    print("Welcome to student management system!")

  # class method
  @classmethod
  def total_student(cls):
    print("Total Student :",cls.count)


# --------------composition-----------------
class StudentManagementSystem:
  def __init__(self):
    self.students = []

  def add_student(self,roll,name,marks):
    s = Student(roll,name,marks)
    self.students.append(s)
    print("student is addded successfully!")

  def view_student(self):
    if not self.students:
      print("No student found!")
    else:
      for s in self.students:
        print(s.show_details())

  def search_student(self,roll):
    for s in self.students:
      if s._Student__roll == roll:
        print(s.show_details())
        return
      print("Student is not found")

  def update_student(self,roll,marks):
    for s in self.students:
      if s._Student__roll == roll:
        s.marks = marks
        print("Studnet marks is updated successfully")
        return
      print("Student is not found")

  def delete_student(self,roll):
    for s in self.students:
      if s._Student__roll == roll:
        self.students.remove(s)
        print("Student is deleted successfully!")
        return
      print("Student is not found")


# --------------main program-------------
Student.greet()
sms = StudentManagementSystem()

while True:
  print("1.Add student")
  print("2.View student")
  print("3.Search student")
  print("4.Update student")
  print("5.Delete student")
  print("6.Total student")
  print("7.Add marks of two students")
  print("8.Exit")

  choice = input("Enter your choice:")
  if choice == "1":
    roll = int(input("Enter roll number"))
    name = input("Enter name")
    marks = int(input("Enter marks"))
    sms.add_student(roll,name,marks)

  elif choice == "2":
    sms.view_student()

  elif choice == "3":
    roll = int(input("Enter roll number to search:"))
    sms.search_student(roll)

  elif choice == "4":
    roll = int(input("Enter roll number to update:"))
    marks = int(input("Enter new marks:"))
    sms.update_student(roll,marks)


  elif choice == "5":
    roll = int(input("Enter roll number to delete:"))
    sms.delete_student(roll)

  elif choice == "6":
    Student.total_student()


  elif choice == "7":
    if len(sms.students)<=2:
      print("sum of first two students' marks:", sms.students[0]+sms.students[1])
    else:
      print("Need at least two students to add marks")

  elif choice == "8":
     print("Exit Program")
     break
  else:
    print("Invalid choice")
```
